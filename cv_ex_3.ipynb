```python
import cv2
import numpy as np
from google.colab.patches import cv2_imshow  # For displaying images in Colab

# ✅ Change this to your uploaded file name
image_path = "jonny depp.jpg"  # Replace with actual image name

# ✅ Step 1: Load the Image
image = cv2.imread(image_path)

# ✅ Step 2: Convert Image to Grayscale
gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

# ✅ Step 3: Load OpenCV's Pre-trained Face Detector
face_cascade = cv2.CascadeClassifier(cv2.data.haarcascades + "haarcascade_frontalface_default.xml")

# ✅ Step 4: Show the Original Image Before Detection
print("🔹 Before Face Detection:")
cv2_imshow(image)

# ✅ Step 5: Detect Faces
faces = face_cascade.detectMultiScale(gray, scaleFactor=1.1, minNeighbors=5, minSize=(50, 50))

# ✅ Step 6: Draw a Rectangle Around the Detected Face
for (x, y, w, h) in faces:
    cv2.rectangle(image, (x, y), (x + w, y + h), (0, 255, 0), 3)  # Green Box

# ✅ Step 7: Show the Image After Detection
print("✅ After Face Detection:")
cv2_imshow(image)

cv2.waitKey(0)
cv2.destroyAllWindows()
```
